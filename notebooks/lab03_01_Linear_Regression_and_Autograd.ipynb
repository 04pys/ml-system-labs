```python
#선형 회귀와 자동 미분에 관한 세션이다
```

```python
import torch
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
#pytorch로 선형회귀 구현하기
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
torch.manual_seed(1)
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
print(x_train)
print(x_train.shape)
```

```python
print(y_train)
print(y_train.shape)
```

```python
# 가중치 W를 0으로 초기화하고 학습을 통해 값이 변경되는 변수임을 명시함.
#torch.zeros() : 첫번째 인자로 크기 받음(아래 경우 크기가 1인 벡터)
W = torch.zeros(1, requires_grad=True)
print(W)
print(W.shape)
```

```python
b = torch.zeros(1, requires_grad=True)
print(b)
print(b.shape)
```

```python
hypothesis = x_train * W + b
print(hypothesis)
```

```python
#torch.mean으로 손실함수(평균제곱오차) 구함
cost = torch.mean((hypothesis - y_train) ** 2)
print(cost)
```

```python
#경사하강법의 일종인 SGD 이용해 경사 하강법 구현
optimizer = optim.SGD([W, b], lr = 0.01) # learning rate는 0.01로
```

```python
# gradient를 0으로 초기화
optimizer.zero_grad()
# 비용 함수를 미분하여 gradient 계산
cost.backward()
# W와 b를 업데이트
optimizer.step()
```

```python
nb_epochs = 1999 # 원하는만큼 경사 하강법을 반복
```

```python
for epoch in range(nb_epochs + 1):
  #H(x) 계산
  hypothesis = x_train * W + b

  #cost 계산
  cost = torch.mean((y_train - hypothesis) ** 2)

  #cost로 H(x) 개선
  optimizer.zero_grad()
  cost.backward() # cost에서 W와 b로
  optimizer.step()

  # 100번마다 로그 출력
  if epoch % 100 == 0:
      print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f} Cost: {:.6f}'.format(
          epoch, nb_epochs, W.item(), b.item(), cost.item()
      ))
```

```python
# 파이토치는 미분을 통해 얻은 기울기를 이전에 계산된 기울기 값에 누적시키는 특징이 있습니다.
# 그렇기때문에
```

```python
w = torch.tensor(2.0, requires_grad=True)

nb_epochs = 20
for epoch in range(nb_epochs + 1):

  z = 2*w

  z.backward()
  print('수식을 w로 미분한 값 : {}'.format(w.grad))
```

```python
w = torch.tensor(2.0, requires_grad=True)
```

```python
y = w**2
z = 2*y + 5
```

```python
z.backward()
```

```python
print('수식을 w로 미분한 값 : {}'.format(w.grad)) # w로 미분하면 4w가 나오고, w = 2이므로 w.grad를 출력하면 8이 나옴
```

```python
print('수식을 y로 미분한 값 : {}'.format(y.grad)) # pytorch는 leaf tensor의 gradient만 저장하므로 y로 미분한 값은 None이다
```
