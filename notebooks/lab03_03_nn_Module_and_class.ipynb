```python
#이전 세션까지는 선형 회귀 이해를 위해 hypothesis와 손실 함수를 직접 구현하여 사용하였다.
#이번 세션에서는 pytorch에서 이미 구현된 가설과 손실함수를 가져와서 사용하는법을 배운다
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
torch.manual_seed(1)
```

```python
# 데이터
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
#가져온 모델이 y=2x라는 정답을 잘 찾아낼 수 있도록 함
```

```python
#nn.Linear()는 입력의 차원, 출력의 차원을 인수로 받습니다.
# 모델을 선언 및 초기화. 단순 선형 회귀이므로 input_dim=1, output_dim=1.
model = nn.Linear(1, 1) #하나의 입력에 대해서 하나의 출력을 가지므로, 입력 차원과 출력 차원 모두 1을 인수로 사용하였습니다.
```

```python
print(list(model.parameters())) # 첫번째 값이 w고 두번째 값이 b이다. 초기값이 랜덤 설정됨. 둘다 학습에 사용되므로 requires_grad=True(backward실행시 둘다 각각 미분됨)
```

```python
# optimizer 설정. 경사 하강법 SGD를 사용하고 learning rate를 의미하는 lr은 0.01
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) # model의 parameters인 w,b가 들어간다
```

```python
nb_epochs = 2000
for epoch in range(nb_epochs + 1):
  #H(x) 계산
  prediction = model(x_train) # prediction은 hypothesis와 같은걸 의미한다

  #cost 계산
  cost = F.mse_loss(prediction, y_train) # <== 파이토치에서 제공하는 평균 제곱 오차 함수

  # cost로 H(x) 개선하는 부분(경사하강법)
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()


  if epoch % 100 == 0:
    # 100번마다 로그 출력
    print('Epoch {:4d}/{} Cost: {:.6f}'.format(
        epoch, nb_epochs, cost.item()
    ))
```

```python
# 임의의 입력 4를 선언
new_var =  torch.FloatTensor([[4.0]])
# 입력한 값 4에 대해서 예측값 y를 리턴받아서 pred_y에 저장
pred_y = model(new_var) # forward 연산
# y = 2x 이므로 입력이 4라면 y가 8에 가까운 값이 나와야 제대로 학습이 된 것
print("훈련 후 입력이 4일 때의 예측값 :", pred_y)
```

```python
#학습 후 w와 b 출력
print(list(model.parameters())) #w = 2, b = 0에 가깝게 나오게 된다. 정답에 가까움.
```

```python
# 데이터
x_train = torch.FloatTensor([[73, 80, 75],
                             [93, 88, 93],
                             [89, 91, 90],
                             [96, 98, 100],
                             [73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])
```

```python
# 모델을 선언 및 초기화. 다중 선형 회귀이므로 input_dim=3, output_dim=1.
#input_dim 은 x가 3개이므로 3, y는 하나
#model = (3x1) 크기의 W와 (1x1) 크기의 b를 가지는 선형회귀 모델.
model = nn.Linear(3, 1)
```

```python
print(list(model.parameters())) # 첫번째는 랜덤 w 3개, 2번째는 랜덤 b 한개. 전부 미분에 이용되기에 requires_grad = True
```

```python
#lr = 0.01이면 기울기가 발산함.
#궁금해서 해봄 ㅎㅎ
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
nb_epochs = 2000
for epoch in range(nb_epochs+1):

  #H(x) 계산(hypothesis)
  prediction = model(x_train)
  # model(x_train)은 model.forward(x_train)와 동일함.

  #cost 계산(손실함수)
  cost = F.mse_loss(prediction, y_train)


  # cost로 H(x) 개선하는 부분
  # gradient를 0으로 초기화
  optimizer.zero_grad()
  # 비용 함수를 미분하여 gradient 계산
  cost.backward()
  # W와 b를 업데이트
  optimizer.step()

  if epoch % 100 == 0:
  # 100번마다 로그 출력
    print('Epoch {:4d}/{} Cost: {:.6f}'.format(
        epoch, nb_epochs, cost.item()
    ))
```

```python
# 위 결과는 learning rate를 0.01로 둬서 cost값이 발산함.(cost를 w,b로 미분한 값에 0.01을 곱해서 w,b에서 빼도 원래의 w,b값보다 cost의 순간변화율(미분값)의 절댓값이 커지게 만듦 )
```
