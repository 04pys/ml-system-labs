```python
import numpy as np
```

```python
vec = np.array([1, 2, 3, 4, 5])
print(vec)
```

```python
mat = np.array([[10, 20, 30], [40 ,50 ,60]])
print(mat)
```

```python
print('vec의 타입: ', type(vec))
print('mat의 타입: ', type(mat))
```

```python
print('vec 축의 개수: ', vec.ndim)
print('vec의 크기(shape): ',vec.shape)
```

```python
print('mat 축의 개수: ', mat.ndim)
print('mat의 크기(shape): ',mat.shape)
```

```python
#모든 값이 0인 2 x 3 배열 생성
zero_mat = np.zeros((2,3))
print(zero_mat)
```

```python
#모든 값이 1인 2 x 3 배열 생성
one_mat = np.ones((2, 3))
print(one_mat)
```

```python
# 모든 값이 특정 상수인 배열 생성. 이 경우 7.
same_value_mat = np.full((2,2), 7)
print(same_value_mat)
```

```python
# 대각선 값이 1이고 나머지 값이 0인 2차원 배열을 생성.
eye_mat = np.eye(3)
print(eye_mat)
```

```python
# 임의의 값으로 채워진 배열 생성
random_mat = np.random.random((2, 2))
print(random_mat)
```

```python
# 0부터 9까지
range_vec = np.arange(10)
print(range_vec)
```

```python
# 1부터 9까지 +2씩 적용되는 범위
range_n_step_vec = np.arange(1,10,2)
print(range_n_step_vec)
```

```python
reshape_mat = np.arange(30).reshape((5,6))
print(reshape_mat)
```

```python
mat = np.array([[1, 2, 3], [4, 5, 6]])
print(mat)
```

```python
slicing_mat = mat[0, :]
print(slicing_mat)
```

```python
print(mat[1,:])
```

```python
slicing_mat = mat[:, 1]
print(slicing_mat)
```

```python
print(mat[:,0])
```

```python
mat = np.array([[1, 2], [4, 5], [7,8]])
print(mat)
```

```python
print(mat[1,0])
```

```python
# 7 5 뽑아서 새로운 배열 생성
indexing_mat = mat[[2,1], [0,1]]
print(indexing_mat)
```

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
```

```python
result = x + y
print(result)
```

```python
result = x - y
print(result)
```

```python
result = result * x
print(result)
```

```python
result = result / x
print(result)
```

```python
mat1 = np.array([[1, 2], [3, 4]])
mat2 = np.array([[5, 6], [7,8]])
mat3 = np.dot(mat1, mat2)
print(mat3)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title('test')
plt.plot([1, 2, 3, 4], [2, 4, 8, 6])
plt.show()
```

```python
plt.title('test')
plt.plot([1, 2, 3, 4], [2, 4, 8, 6])
plt.xlabel('hours')
plt.ylabel('scores')
plt.show()
```

```python
plt.title('students')
plt.plot([1,2,3,4],[2,4,8,6])
plt.plot([1.5,2.5,3.5,4.5],[3,5,8,10]) # 라인 새로 추가
plt.xlabel('hours')
plt.ylabel('score')
plt.legend(['A student', 'B student']) # 범례 삽입
plt.show()
```
